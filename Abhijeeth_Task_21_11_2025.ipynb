1. If Else

```python
users = [
    ["abhi", "abhi1234"],
    ["theertharaj", "t245A1"],
    ["john", "john123"]
]
username = input("Enter username: ")
password = input("Enter password: ")
if [username, password] in users:
    print("Login Successful")
else:
    print("Access Denied")
```

2. Tuple

```python
my_tuple = (1, 3, 5, 3, 7, 3, 9)
value = int(input("Enter value to count: "))
count = my_tuple.count(value)
print("Count:", count)
```

3. LISTS

```python
numbers = []
for i in range(5):
    num = int(input("Enter number: "))
    numbers.append(num)
print("List:", numbers)
```

4. VARIABLE

```python
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
a=a+b
b= a-b
a= a-b
print("After swapping:")
print("a =", a)
print("b =", b)
```

5. Sorting


```python
words = ["apple", "developer", "banana", "dog"]
words.sort(key=len)
print("Sorted by length:", words)
```

```python
data = [(10, 3), (11, 23), (8, 4)]
sorted_data = sorted(data, key=lambda x: x[1])
print("Sorted tuples:", sorted_data)
```

```python
nums = [5, 2, 9, 1, 7]
nums.sort()
print("Sorted List:", nums)
```

6. Palindrome

```python
num = int(input("Enter number: "))
if str(num) == str(num)[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
text = input("Enter string: ").lower()
if text == text[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
words = ["madam", "apple", "level", "wow", "chair"]
palindromes = [w for w in words if w == w[::-1]]
print("Palindromes:", palindromes)
```
