```python
a=int(input('enter a number to check odd or even'))
num = a
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

```python
ch=input('enter a character to check whether a letter is vowel or not :')

if len(ch) != 1:
    print("Error: Please enter only ONE character!")
else:
    if ch in "aeiouAEIOU":
        print(ch, "is a Vowel")
    else:
        print(ch, "is a Consonant")
```

```python
status=int(input('Enter web page error code :'))
if 200 <= status < 300:
    msg = "Success"
elif 300 <= status < 400:
    msg = "Redirect"
elif 400 <= status < 500:
    msg = "Client Error"
else:
    msg = "Server Error"
print(msg)
```

### **Tuple**

```python
colors = ("red", "green", "blue")
print(colors)
print(colors[1])
print(colors[0:2])
print(colors.index('blue'))
```

```python
data = (1, 2, 3, 4, 5, 6)
first, *middle, last = data
print(f'First: {first}, Middle: {middle}, Last: {last}')
```

### **palindrome**

```python
s = "madam"
if s == s[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
text = "level"
rev = ""
for i in text:
    rev = i + rev
print("Palindrome" if text == rev else "Not Palindrome")
```

```python
nums = [4, 2, 8, 6]
nums.sort(reverse=True)
print(nums)
```
