## **if-else**

```python
number=int(input('enter a number : '))
#number = 10
if number >= 0:
    print(f"{number} is a positive number.")
else:
    print(f"{number} is a negative number.")
```

```python

value = int(input("Enter a value: "))
if value < 50:
    print(f"{value} is small.")
elif 50 <= value < 100:
    print(f"{value} is medium.")
else:
    print(f"{value} is large.")
```

```python
is_logged_in = True
has_admin_privileges = False

if is_logged_in:
    print("User is logged in.")
    if has_admin_privileges:
        print("User has administrator privileges.")
    else:
        print("User does not have administrator privileges.")
else:
    print("User is not logged in.")
```

# **tuple**

```python
number = ("1", "2", "3")
print(number)
print(number[1])
print(number[0:2])
print(number.index('2'))
```

```python
data = (8, 7, 6, 5, 4, 3, 2)
first, *middle, last = data
print(f'First: {first}, Middle: {middle}, Last: {last}')
```

```python
# Packing
person_info = ("Alice", 30, "New York")
print(f"Packed tuple: {person_info}")
# Unpacking
name, age, city = person_info
print(f"Unpacked name: {name}")
print(f"Unpacked age: {age}")
print(f"Unpacked city: {city}")
```

#  **lists**

```python
my_list = [1, 2, 3, 4]
print(" Created list: ",my_list)
```

```python
my_list.append(6)
print(" Appended 6: {my_list}")
my_list.insert(0, 0)
print(" Inserted 0 at index 0: {my_list}")
print(my_list)
```

```python
sub_list = my_list[1:3]
print(" Sliced list [1:3]: ",sub_list)
```

#  **sorting** **function**

```python
numbers = [3, 1, 4, 1, 5, 9]
sorted_numbers = sorted(numbers)
print(" Original: ",numbers," Sorted with sorted(): ",sorted_numbers)
```

```python
words = ['banana', 'apple', 'cherry']
words.sort()
print(" Sorted words: ",words)
```

```python
numbers_desc = [3, 1, 4, 1, 5, 9]
sorted_desc = sorted(numbers_desc, reverse=True)
print("descending order : ",sorted_desc)
```

```python
pairs = [(1, 'b'), (3, 'a'), (2, 'c')]
sorted_pairs = sorted(pairs)
print("Sorted tuples: ",sorted_pairs)
```

```python
fruits = ['apple', 'banana', 'kiwi', 'grapefruit']
sorted_fruits_by_len = sorted(fruits, key=len)
print("Sorted fruits: ",sorted_fruits_by_len)
```

# **palindrome**

```python
word1 = "madam"
is_palindrome1 = word1 == word1[::-1]
print(" Is ",word1,"a palindrome? ",is_palindrome1)
```

```python
a = "Racecar"
cleaned_phrase1 = a.lower()
is_palindrome3 = cleaned_phrase1 == cleaned_phrase1[::-1]
print("Is ",a ," a palindrome?", is_palindrome3)
```
